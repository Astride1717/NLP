```python
import re
def cleanText(data): 
    #텍스트에 포함되어 있는 특수 문자 제거 
    text1 = re.sub('[ -_=+,#/\?:^$.@*\"※~&%ㆍ!』<>]','', data) #특수문자 제거
    p = re.compile("[^0-9]") #숫자 제거
    done = text1.replace(" ", "") #공백제거
    #done = "".join(p.findall(text1))
    return done
```

```python
#N.B : in구문에 꼭 리스트가 들어가야 에러가 안난다는것
#시리즈를 넣으면 각 줄(리스트)가 하나의 문자열로 인식돼 잘못 출력됨 )
#모델링을 했을때 다 없앤 nan가 토픽주요단어로 뜨는 문제
words1 = []
for x in range(len(tokenized_mecab)):
    y = [cleanText(i) for i in tokenized_mecab[x] if i not in stopwords and i not in ['nan','\xa0', ''] and not pd.isnull(i)]
    words1.append(y)
```
