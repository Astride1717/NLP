```python
#불용어 처리
stop = pd.read_csv('/mnt/c/Users/NY/Anaconda3/DataScience/따릉따릉/따릉이 민원/불용어_리스트.txt', encoding = 'utf-8', header = None, sep = "	" )
```

```python
stopwords = stop[0].values
stopwords
```

```python
#특수문자/숫자 제거
import re
 
def cleanText(data):
    text = re.sub('[_{}\n;-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', ' ', data)
    text1 = text.lower()
    p = re.compile("[^0-9]")
    text2 = "".join(p.findall(text1))
 
    return text2
```

```python
cleaned = train.data.apply(cleanText)
```

```python
cleaned
```

```python
from konlpy.tag import Mecab
m = Mecab()
```

```python
m.morphs('셀레니움')
```

```python
text = m.morphs(cleanText(train.data[3]))
print(text)
```

```python
tokenized_mecab = []
for i in range(len(cleaned)):
    tokens = m.morphs(cleaned[i])
    tokenized_mecab.append(tokens)
```

```python
tokenized_mecab
```

```python
words1 = []
for x in range(len(tokenized_mecab)):
    y = [i for i in tokenized_mecab[x] if i not in stopwords and i not in ['  ',' ', '']]
    words1.append(y)
```

```python
train["preprocessed"] = pd.Series(words1)
```

```python
train["preprocessed"]
```

```python
detokenized_doc = []
for i in range(len(train.data)):
    t = ' '.join(words1[i])
    detokenized_doc.append(t)
```

```python
train['detokenized'] = pd.Series(detokenized_doc)
```
